Sentiment analysis

```python
from transformers import pipeline

classifier = pipeline('sentiment-analysis')
result = classifier("I was not happy with the last movie")
print(result)
```

```python
pipeline(task = 'sentiment-analysis')('i was not happy')
```

```python
pipeline(task = 'sentiment-analysis', model = 'facebook/bart-large-mnli')("the movie was good but some parts seemed off, it was too long movie which made me bored")
```

Batch sentiment analysis

```python
classifier = pipeline('sentiment-analysis')
text_list = ['i really liked that movie',
             'i am unhappy with the product',
             'the food was tasty but there were too spices which made it very spicy']
classifier(text_list)
```

Emotion and sentiment analysis with a strong model

```python
classifier = pipeline('sentiment-analysis', model= 'SamLowe/roberta-base-go_emotions')
text_list = ['i really liked that movie',
             'i am unhappy with the product',
             'the food was tasty but there were too spices which made it very spicy']
classifier(text_list)
```

Text generation

```python
text_generator = pipeline('text-generation', model = 'distilbert/distilgpt2')
generated_text = text_generator("Today is a rainy day in london",
                                truncation = True,   #This parameter is typically used when the input text might be longer than the maximum sequence length the model can handle. When set to True, the input text will be truncated (cut short) to fit within the model's maximum input size'''
                                num_return_sequences = 2) #simply tells the text generation model how many different text outputs it should create for your given input.
print(generated_text[0]['generated_text'])
```

Question Answering

```python
qa_model = pipeline('question-answering')
question = "what is my job?"
context = "I am developing AI models with python"
answer = qa_model(question = question, context = context)
print(answer)
```
