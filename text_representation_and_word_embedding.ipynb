```python
import pandas as pd
import numpy as np
```

Bag of words

```python
df = pd.DataFrame({"text":['people watch ds',
                           'ds watch ds',
                           'people write comment',
                           'ds write comment'],"output":[1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow.toarray())
```

BIgram BOW

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow.toarray())
```

Tigram BOW

```python
cv = CountVectorizer(ngram_range=(3,3))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow.toarray())
```
